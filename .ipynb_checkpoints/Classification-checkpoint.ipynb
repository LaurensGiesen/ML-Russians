# 2 Train/val/test-split and standardisation

1. Splitting the data

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df.drop(['outcome_damage_inc', 'outcome_profit', 'outcome_damage_amount'], 1)
y = df['outcome_damage_inc']

X_train_val_inc, X_test_inc, y_train_val_inc, y_test_inc = train_test_split(X, y, test_size=0.2, random_state=1234)
X_train_inc, X_val_inc, y_train_inc, y_val_inc = train_test_split(X_train_val_inc,y_train_val_inc, test_size=0.2, random_state=1234)
```

__note__: Standardization already happened in pre-processing

# 3. Logistic regression

3.1 Basic model

1. Train a logistic model

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(C=10000, max_iter=200) 
logreg.fit(X_train_inc, y_train_inc)
```

uitleg

2. Test the train dataset

```python
print(logreg.score(X_train_inc, y_train_inc))
```

uitleg

3. Evaluate the model

```python
# predict the outcomes for the train dataset
y_train_pred = logreg.predict(X_train_inc)
y_train_pred_prob = logreg.predict_proba(X_train_inc)
```

```python
# construct the confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_train_inc, y_train_pred)
print(confusion_matrix)
```

We evalueren ons model door te kijken naar confusion matrix

```python
# print the classification report
from sklearn.metrics import classification_report
print(classification_report(y_train_inc, y_train_pred))
```

We evalueren ons model door te kijken naar accuracy + classification report

```python
# print the ROC-curve
import sklearn.metrics as metrics
preds = y_train_pred_prob[:,1]
fpr, tpr, threshold = metrics.roc_curve(y_train_inc, preds)
roc_auc = metrics.auc(fpr, tpr)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

We evalueren ons model door te kijken naar de ROC-curve

2. Calibrate model with the isotonic calibration

3. Construct the classification report for the calibrated model

```python
y_pred = logreg.predict(X_train_inc)
y_pred_c = isotonic.predict(X_train_inc)

print(classification_report(y_train_inc, y_pred))
print(classification_report(y_train_inc, y_pred_c))
```

### 3.3 Adding polynomials and penalisation

1. Design the polynomial features

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=1) #die 1 mag eventueel hoger
X_train_poly_inc = poly.fit_transform(X_train_inc)
X_train_val_poly_inc = poly.transform(X_train_val_inc)
```

2. Do the RandomizedSearch

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn import metrics

C = [round(x,5) for x in np.linspace(start = 0.0001, stop = 10, num = 1000)]
random_grid = {'C': C}

logreg = LogisticRegression( max_iter=1000) 

log_random = RandomizedSearchCV(estimator = logreg, param_distributions = random_grid, n_iter = 100,
                               cv = 3, verbose=4,  n_jobs=-1, random_state=1234)


log_random.fit(X_train_val_poly_inc, y_train_val_inc)
```

3. Look at the parameters of the best model 

```python
log_random.best_params_
```

4. Make a scatterplot of the accuracy vs the C-parameter

```python
out2 = pd.DataFrame(log_random.cv_results_)
plt.ylim(out2['mean_test_score'].min()-0.0001, out2['mean_test_score'].max()+0.0001)
plt.scatter(out2['param_C'], out2['mean_test_score'], c='blue');
```

hier kijken welke C we willen meegeven aan de Grid search

5. Grid search

```python
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
from sklearn.metrics import make_scorer

prec_scorer = make_scorer(metrics.accuracy_score)
# other possibilities are metrics.recall_score, metrics.average_precision, ...

C = [x for x in np.linspace(start = 0.0001, stop = 1, num = 50)]

random_grid = {'C': C}

# The object to fit the model
logreg_balanced = LogisticRegression( max_iter=1000) 

# object for the randomised search
log_grid = GridSearchCV(estimator = logreg_balanced, param_grid = random_grid,
                                scoring=prec_scorer  , cv = 3, verbose=10,  n_jobs=-1)
isotonic_log = CalibratedClassifierCV(log_grid, cv=5, method='isotonic')

#Fitting model (model+ calibration)
isotonic_log.fit(X_train_val_poly_inc, y_train_val_inc)
```

6. Evaluation

```python
isotonic_log.score(X_train_poly_inc, y_train_inc)
```

```python
y_pred_poly = isotonic_log.predict(X_train_poly_inc)
print(classification_report(y_train_inc, y_pred_poly))
```

```python
import sklearn.metrics as metrics
# calculate the fpr and tpr for all thresholds of the classification
preds = y_train_pred_prob[:,1]
preds_poly = isotonic_log.predict_proba(X_train_poly_inc)[:,1]

fpr, tpr, threshold = metrics.roc_curve(y_train_inc, preds)
fpr_p, tpr_p, threshold = metrics.roc_curve(y_train_inc, preds_poly)

roc_auc = metrics.auc(fpr, tpr)
roc_auc_p = metrics.auc(fpr_p, tpr_p)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'logistic: AUC =  %0.2f' % roc_auc)
plt.plot(fpr_p, tpr_p, 'g', label = 'polynomial AUC = %0.2f' % roc_auc_p)

plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

3. Deal with the unbalance

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline

over = SMOTE(sampling_strategy=0.8, random_state=1703)
under = RandomUnderSampler(sampling_strategy=1)
pipeline = Pipeline([('o', over), ('u', under)])

X_train_smote_inc, y_train_smote_inc = pipeline.fit_resample(X_train_inc.copy(), y_train_inc.copy())
X_train_val_smote_inc, y_train_val_smote_inc = pipeline.fit_resample(X_train_val_inc.copy(), y_train_val_inc.copy())
np.bincount(y_train_val_smote_inc)
```

4. Redo the grid search, but on the balanced data and calibrate the model

```python
n_neighbors = 4*np.arange(1,50)
param_grid = {'n_neighbors': n_neighbors}

knn_b = KNeighborsClassifier( )
grid_search_knn = GridSearchCV(estimator = knn, param_grid = param_grid, scoring=prec_scorer ,cv = 3,  verbose=5, n_jobs = -1)
isotonic_knn = CalibratedClassifierCV(grid_search_knn, cv=3, method='isotonic')


isotonic_knn.fit(X_train_val_smote_inc, y_train_val_smote_inc)
```

5. Re-evaluate

```python
y_pred_knn_b = isotonic_knn.predict(X_train_inc)
print(classification_report(y_train_inc, y_pred_knn_b))
```

```python
preds_knn_b = isotonic_knn.predict_proba(X_train_inc)[:,1]

fpr, tpr, threshold = metrics.roc_curve(y_train_inc, preds)
fpr_p, tpr_p, threshold = metrics.roc_curve(y_train_inc, preds_poly)
fpr_k, tpr_k, threshold = metrics.roc_curve(y_train_inc, preds_knn_b)

roc_auc = metrics.auc(fpr, tpr)
roc_auc_p = metrics.auc(fpr_p, tpr_p)
roc_auc_k = metrics.auc(fpr_k, tpr_k)


plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'logistic: AUC =  %0.2f' % roc_auc)
plt.plot(fpr_p, tpr_p, 'g', label = 'polynomial AUC = %0.2f' % roc_auc_p)
plt.plot(fpr_k, tpr_k, 'r', label = 'knn AUC = %0.2f' % roc_auc_k)

plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

# 2. GBM

1. Randomized search

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import RandomizedSearchCV


n_estimators = [int(x) for x in np.linspace(start = 0, stop = 200, num = 50)]
max_features = ['auto', 'sqrt', 'log2']
max_depth = [int(x) for x in np.linspace(1, 10, num = 10)]
min_samples_split = [int(x) for x in np.linspace(2, 10, num = 9)]
min_samples_leaf = [int(x) for x in np.linspace(1, 10, num = 10)]
learning_rate = [round(x,5) for x in np.linspace(start = 0.00001, stop = 1.5, num = 50)]

random_grid = {'n_estimators': n_estimators,
               'learning_rate': learning_rate,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
gbm = GradientBoostingClassifier()
gbm_random = RandomizedSearchCV(estimator = gbm, param_distributions = random_grid, n_iter = 300, cv = 5, verbose=3, random_state=4872, n_jobs = -1)
gbm_random.fit(X_train_smote_inc, y_train_smote_inc)
```

```python
gbm_random.best_params_
```

2. Evaluate

```python
y_pred_gbm = gbm_random.predict(X_train_inc)
print(classification_report(y_train_inc, y_pred_gbm))
```

3. scatterplot

```python
out2 = pd.DataFrame(gbm_random.cv_results_)

xlabel_names = ['param_max_depth','param_min_samples_split','param_min_samples_leaf','param_n_estimators',
                'param_learning_rate','param_max_features']

fig, axs = plt.subplots(2,3, figsize=(20,10))

axs[0,0].scatter(out2['param_max_depth'], out2['mean_test_score'], c='blue');
axs[0,0].set_title('max_depth')

axs[0,1].scatter(out2['param_min_samples_split'], out2['mean_test_score'], c='blue');
axs[0,1].set_title('min_samples_split')

axs[0,2].scatter(out2['param_min_samples_leaf'], out2['mean_test_score'], c='blue');
axs[0,2].set_title('min_samples_leaf')

axs[1,0].scatter(out2['param_n_estimators'], out2['mean_test_score'], c='blue');
axs[1,0].set_title('n_estimators')

axs[1,1].scatter(out2['param_learning_rate'], out2['mean_test_score'], c='blue');
axs[1,1].set_title('learning_rate')

axs[1,2].scatter(out2['param_max_features'], out2['mean_test_score'], c='blue');
axs[1,2].set_title('max_features')

for ax in axs.flat: ax.set(ylabel='r_squared')
```

4. Grid search and calibration

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.calibration import CalibratedClassifierCV, calibration_curve

n_estimators = [89]
learning_rate = [0.00001]
max_features = ['log2', 'sqrt']
max_depth = [8,7]
min_samples_split = [5,9,10]
min_samples_leaf = [7,4]
                                            
random_grid = {'n_estimators': n_estimators,
               'learning_rate': learning_rate,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
gbm = GradientBoostingClassifier()
gbm_grid = GridSearchCV(estimator = gbm, param_grid = random_grid,  cv = 5, verbose=10,  n_jobs = -1)

isotonic_gbm = CalibratedClassifierCV(gbm_grid, cv=5, method='isotonic')
isotonic_gbm.fit(X_train_val_smote_inc, y_train_val_smote_inc)
```

5. Evaluate
