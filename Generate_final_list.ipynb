# Generate a list of 200 applicants

# 1. Loading packages and dataset

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-darkgrid')

df = pd.read_csv('score_with_all_results.csv', index_col=[0])
```

We sort our list by 'outcome_profit' - 'outcome_profit_damage' so we see who will realize the most profit for the hotel.

```python
df['profit_after_damage_reduction'] = (df['outcome_profit']-df['outcome_damage_amount'])
df.sort_values(by = ['profit_after_damage_reduction'], ascending=False)
```

No we take the first 200 people as past applicants

```python
passed_applicants = df.iloc[:199,:]
passed_applicants.to_csv('final_list.csv')
```
